```python
import speech_recognition as sr
from gtts import gTTS
import os
import pyjokes
import wikipedia
from datetime import datetime
import webbrowser
from IPython.display import Audio, display
from pydub import AudioSegment
from pydub.playback import play
import tempfile
import pygame
```

```python
# Texto para fala (Text-to-Speech)
def speak(text, lang='en'):
    tts = gTTS(text=text, lang=lang)
    temp_path = "/tmp/voice.mp3"
    tts.save(temp_path)
    display(Audio(temp_path, autoplay=True))

# Fala para texto (Speech-to-Text)
def get_audio():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("üé§ Aguardando sua fala...")
        recognizer.adjust_for_ambient_noise(source, duration=1)
        audio = recognizer.listen(source)

    try:
        print("üß† Reconhecendo...")
        text = recognizer.recognize_google(audio)
        print(f"Voc√™ disse: {text}")
        return text.lower()
    except sr.UnknownValueError:
        speak("Desculpe, n√£o entendi.")
    except sr.RequestError:
        speak("Servi√ßo de reconhecimento n√£o est√° dispon√≠vel.")
    return ""
```

```python
# Fun√ß√£o de resposta
def respond(command):
    print("üìå Comando recebido:", command)

    if 'youtube' in command:
        speak("O que voc√™ deseja buscar no YouTube?")
        keyword = get_audio()
        if keyword:
            url = f"https://www.youtube.com/results?search_query={keyword}"
            webbrowser.open(url)
            speak(f"Aqui est√£o os resultados para {keyword} no YouTube.")

    elif 'search' in command or 'wikipedia' in command:
        speak("O que voc√™ deseja saber?")
        query = get_audio()
        if query:
            try:
                result = wikipedia.summary(query, sentences=2)
                speak("De acordo com a Wikip√©dia:")
                print(result)
                speak(result)
            except wikipedia.exceptions.DisambiguationError:
                speak("Esse termo √© amb√≠guo. Seja mais espec√≠fico, por favor.")
            except wikipedia.exceptions.PageError:
                speak("N√£o encontrei nada sobre isso na Wikip√©dia.")

    elif 'joke' in command or 'piada' in command:
        joke = pyjokes.get_joke(language='en')
        print(joke)
        speak(joke)

    elif 'what time' in command or 'que horas' in command:
        time_now = datetime.now().strftime("%H:%M")
        speak(f"Agora s√£o {time_now}")

    elif 'farm√°cia' in command or 'pharmacy' in command:
        speak("Abrindo farm√°cias pr√≥ximas no Google Maps.")
        webbrowser.open("https://www.google.com/maps/search/farm√°cia+mais+pr√≥xima")

    elif 'stop' in command or 'sair' in command or 'exit' in command:
        speak("Encerrando o assistente. At√© logo!")
        raise SystemExit
```

```python
from gtts import gTTS
from IPython.display import Audio, display, HTML
import base64

def gerar_e_exibir_audio(texto, lang, nome_arquivo, slow=False):
    caminho = f"/tmp/{nome_arquivo}.mp3"
    tts = gTTS(text=texto, lang=lang, slow=slow)
    tts.save(caminho)

    # Player
    display(Audio(caminho, autoplay=True))

    # Download button
    with open(caminho, "rb") as f:
        b64 = base64.b64encode(f.read()).decode()
    html = f"""
    <a download="{nome_arquivo}.mp3" href="data:audio/mp3;base64,{b64}" target="_blank">
        üì• Baixar {nome_arquivo}.mp3
    </a>
    """
    display(HTML(html))

# Testes
gerar_e_exibir_audio("How are you doing?", lang='en', nome_arquivo="english")
gerar_e_exibir_audio("Je vais au supermarch√©", lang='fr', nome_arquivo="french", slow=True)
```

```python
# Instala√ß√£o de depend√™ncias (roda no Colab)
!pip install pydub speechrecognition ffmpeg-python --quiet

from google.colab import files
from IPython.display import Audio, display
from pydub import AudioSegment
import speech_recognition as sr
import os

# Upload do arquivo
print("üì§ Envie um arquivo de √°udio (mp3, wav, ogg...)")
uploaded = files.upload()

# Pega o nome do arquivo
for filename in uploaded.keys():
    print(f"üìÅ Arquivo recebido: {filename}")

    # Converte para WAV
    if not filename.endswith(".wav"):
        audio = AudioSegment.from_file(filename)
        filename_wav = filename.rsplit(".", 1)[0] + ".wav"
        audio.export(filename_wav, format="wav")
        print(f"üîÑ Convertido para WAV: {filename_wav}")
    else:
        filename_wav = filename

    # Toca o √°udio
    display(Audio(filename_wav))

    # Reconhecimento de fala
    recognizer = sr.Recognizer()
    with sr.AudioFile(filename_wav) as source:
        audio_data = recognizer.record(source)

    try:
        texto = recognizer.recognize_google(audio_data, language='pt-BR')
        print("üìù Texto reconhecido:")
        print(texto)
    except sr.UnknownValueError:
        print("‚ö†Ô∏è N√£o foi poss√≠vel entender o √°udio.")
    except sr.RequestError as e:
        print(f"‚ùå Erro ao se comunicar com o servi√ßo de reconhecimento: {e}")
```
